```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
years = 30
initial = 10000

sjp_return = 0.05
index_return = 0.068
```

```python
data = []
```

```python
sjp_value = initial
index_value = initial
```

```python
monthly = 200

for year in range(years + 1):
    difference = index_value - sjp_value

    data.append([year, sjp_value, index_value, difference])

    sjp_value = sjp_value * (1 + sjp_return) + (monthly * 12)
    index_value = index_value * (1 + index_return) + (monthly * 12)
```

```python
import pandas as pd

df = pd.DataFrame(data, columns=["Year", "SJP", "Index Fund", "Difference"])

df.head()
```

```python
plt.plot(df["Year"], df["Difference"], label="Wealth Lost to Fees")

plt.xlabel("Years")
plt.ylabel("Difference (£)")
plt.title("Wealth Lost to Fees Over Time")

plt.legend()
plt.show()
```

```python
plt.plot(df["Year"], df["SJP"], label="SJP (5%)")
plt.plot(df["Year"], df["Index Fund"], label="Index Fund (6.8%)")

plt.xlabel("Years")
plt.ylabel("Portfolio Value (£)")
plt.title("Impact of Fees on Investment Growth")

plt.legend()
plt.show()
```
